## 2-Stock Simulation
**Objective:** Simulate and visualize price paths for 2 stocks.

**Key Concepts Learned:**

## 5-Stock Simulation
**Objective:** Extend simulation to 5 stocks with different volatilities.

**Key Concepts Learned:**
- Looping through multiple stocks
- Storing stock parameters in a dictionary
- Plotting multiple lines on the same graph


**2 Stock Simulation**

```
import numpy as np
import matplotlib.pyplot as plt

# parameters
days = 100
price = 100
returns = np.random.normal(0, 0.01, days)   # mean=0%, std=1%
returnss = np.random.normal(0, 0.02, days)
# simulate prices
price_path = price * np.cumprod(1 + returns)
pricepath=price * np.cumprod(1 + returnss)

# Create an array for the x-axis (days)
x_days = np.arange(days)

# plot
plt.plot(x_days, price_path, label='Stock A', color='red')
plt.plot(x_days, pricepath, label='Stock B', color='blue')
plt.title("Simulated Stock Price Path")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.show()
```

**4 or 5 stocks randomly generated to compare**


```
import numpy as np
import matplotlib.pyplot as plt

# Parameters
days = 100
stocks = {
    "Stock A": {"price": 100, "vol": 0.01},
    "Stock B": {"price": 120, "vol": 0.02},
    "Stock C": {"price": 90, "vol": 0.015},
    "Stock D": {"price": 150, "vol": 0.012},
    "Stock E": {"price": 80, "vol": 0.018}
}

# Prepare x-axis
x_days = np.arange(1, days+1)

# Simulate price paths
for name, params in stocks.items():
    returns = np.random.normal(0, params["vol"], days)
    price_path = params["price"] * np.cumprod(1 + returns)
    plt.plot(x_days, price_path, label=name)

# Plot formatting
plt.title("Simulated Stock Price Paths (5 Stocks)")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.show()
```
